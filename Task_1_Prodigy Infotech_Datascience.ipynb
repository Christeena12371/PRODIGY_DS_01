```python
#Dropping unnecesary columns from the dataset 
df1=df.drop(['Country Code', 'Indicator Name', 'Indicator Code'],axis=1)
df1
```

```python
# Histogram plot for years
years=['1960','1961','1962','1963','1964','1965','1966','1967','1968','1969','1970','1971','1972','1973','1974','1975','1976',
       '1977','1978','1979','1980','1981','1982','1983','1984','1985','1986','1987','1988','1989','1990','1991','1992','1993',
       '1994','1995','1996','1997','1998','1999','2001','2002','2003','2004','2005','2006','2007','2008','2009','2010','2011',
       '2012','2013','2014','2015','2016','2017','2018','2019','2020','2021','2022']
plt.figure(figsize=(33,6))
plt.hist(years)
plt.show()
```

```python
# Plot for the maximum population of top 10 countries in the year 1960
max_1960=df1.sort_values(by='1960',ascending=True).head(10)

plt.figure(figsize=(10,3))
plt.bar(max_1960['Country Name'],max_1960['1960'],color='blue')
plt.title('Maximum population of top 10 countries in the year 1960')
plt.xlabel('Country')
plt.ylabel('Population')
plt.xticks(rotation=90)

#Similarly we can plot the maximum poulation of top 10 countries of other years.
```

```python
# Plot for the minimum population of top 10 countries in the year 1960
min_1960=df1.sort_values(by='1960',ascending=True).head(10)

plt.figure(figsize=(10,3))
plt.bar(min_1960['Country Name'],min_1960['1960'],color='blue')
plt.title('Population of top 10 countries in the year 1960')
plt.xlabel('Country')
plt.ylabel('Population')
plt.xticks(rotation=90)


#Similarly we can plot the minimum population of top 10 countries for other years.
```

```python
# Plot for the year which has maximum population
max_population_year=df1.loc[:,'1960':'2022'].idxmax(axis=1)

#Plot for the year with maximum population for each country
no_of_years=max_population_year.value_counts()

plt.figure(figsize=(10,4))
no_of_years.plot(kind='bar',color='blue')
plt.title('Year with maximum population for each country')
plt.xlabel('year')
plt.ylabel('Number of countries')

plt.show()
```

```python
#Plot for the year wise mean population for each year
mean_population=df.iloc[:,4:].mean(axis=0)

plt.figure(figsize=(20,4))
mean_population.plot(kind='bar',color='blue')
plt.title('Mean population')
plt.xlabel('Year')
plt.ylabel('No.of poulation')
```
